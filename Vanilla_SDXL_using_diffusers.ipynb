# SDXL Base 1.0

```python
from diffusers import DiffusionPipeline
import torch

pipe = DiffusionPipeline.from_pretrained("stabilityai/stable-diffusion-xl-base-1.0", torch_dtype=torch.float16, use_safetensors=True, variant="fp16")
pipe.to("cuda")
```

```python
# if using torch < 2.0
pipe.enable_xformers_memory_efficient_attention()
```

```python
prompt = "photography, realistic, 1girl, 1guy, Hardcore Sex, flirty, Erotic orgasm pleasure, light damp skin, Bed, (Sex Positions: doggy) , (skin pores, veins, moles, freckles:0.5) , Voluptuous"

image = pipe(prompt=prompt, num_inference_steps=1, guidance_scale=1).images[0]

image
```

```python
import random
import string

def generate_random_string(length=10):
    """Generates a random string of letters and digits."""
    letters_and_digits = string.ascii_letters + string.digits
    return ''.join(random.choice(letters_and_digits) for _ in range(length))
```

```python
from PIL import Image, ImageDraw, ImageFont
import textwrap

def add_prompt_strip(image, prompt_text):
    """Adds a black strip at the bottom of the image with the prompt text."""
    width, height = image.size
    strip_height = 100  # Adjust as needed
    new_height = height + strip_height

    # Create a new image with extra space for the strip
    new_image = Image.new("RGB", (width, new_height), color="white")
    new_image.paste(image, (0, 0))

    # Draw the black strip
    draw = ImageDraw.Draw(new_image)
    draw.rectangle((0, height, width, new_height), fill="black")

    # Add the prompt text

    # font = ImageFont.load_default()
    font_size = 30

    font = ImageFont.truetype("arial.ttf", font_size)

    # Wrap the text
    lines = textwrap.wrap(prompt_text, width=40)  # Adjust width as needed

    # Calculate text height based on number of lines and font size
    text_height = len(lines) * (font_size + 5) # Add 5 for line spacing (adjust as needed)

    # Calculate strip height to fit wrapped text
    strip_height = text_height + 20  # Add padding (adjust as needed)

    # Create a new image with extra space for the strip
    width, height = image.size
    new_height = height + strip_height
    new_image = Image.new("RGB", (width, new_height), color="white")
    new_image.paste(image, (0, 0))

    # Draw the black strip
    draw = ImageDraw.Draw(new_image)
    draw.rectangle((0, height, width, new_height), fill="black")

    # Draw wrapped text lines
    y_text = height + 10  # Start position for first line (adjust as needed)
    for line in lines:
        text_width, _ = font.getbbox(line)[2:] # Get text width
        text_x = (width - text_width) // 2  # Center text horizontally
        draw.text((text_x, y_text), line, fill="white", font=font)
        y_text += font_size + 5  # Move to next line (adjust spacing as needed)

    return new_image

image_with_strip = add_prompt_strip(image, prompt)
image_with_strip.save(generate_random_string() + ".png")
image_with_strip
```
